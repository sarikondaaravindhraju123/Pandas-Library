```python
# convert day[string] into date:
df = pd.read_csv("weather_data1.csv",parse_dates=["day"])
df
```

```python
type(df.day[0])
```

```python
df
```

```python
# change the index as day
df.set_index("day",inplace=True)
df
```

```python
# Replace NaN values with 0
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    "temperature" : 0,
    "winspeed" : 0,
    "event" : "no_event"
})
new_df
```

```python
# ffill and bfill both are used for estimated values.
# It defaultly copies value by vertically
# If we change the axis then it will change
```

```python
# ffill : It fill null values with upper row values
new_df = df.fillna(method="ffill")
new_df
```

```python
# bfill: It fills null value with lower row values:
new_df = df.fillna(method="bfill")
new_df
```

```python
# change the axis:
new_df = df.fillna(method="bfill",axis="columns")
new_df
```

```python
# limit: limits the maximum no of consecutive missing values to be filled to 1
new_df = df.fillna(method="ffill",limit=1)
new_df
```

```python
df
```

```python
new_df = df.fillna(method="ffill",limit=2)
new_df
```

```python
new_df = df.fillna(method="bfill",limit=2)
new_df
```

```python
# interpolate: It returns the values linearly
# It is average of upper and lower column
# when there are 3 NaN values there will be 0.25 difference
# interpolate defaultly will be linear
```

```python
df
```

```python
new_df = df.interpolate()
new_df
```

```python
# dropna : this drop rows having NaN values
new_df = df.dropna()
new_df
```

```python
df.set_index("day",inplace=True)
df
```

```python
# drop rows if every column has null values:
new_df = df.dropna(how="all")
new_df
```
