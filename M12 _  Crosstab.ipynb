```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("survey.xls")
df
```

```python
# cross tab is on pandas not on dataframe
# first one consists of x-axis & second one consists of y-axis
# The values we see is the one occur how many  time

pd.crosstab(df.Nationality,df.Handedness)
```

```python
# Display sex and total of columns and rows:
# Margin : Displays total of columns and rows

pd.crosstab(df.Sex,df.Handedness,margins=True)
```

```python
pd.crosstab(df.Sex,[df.Handedness,df.Nationality],margins=True)
```

```python
pd.crosstab([df.Sex,df.Nationality],df.Handedness,margins=True)
```

```python
# Instead of values get percentage:
# Hint : Normalize

pd.crosstab(df.Sex,df.Handedness,normalize=True)
```

```python
# Finding average of people according to the nation:

pd.crosstab(df.Sex,df.Handedness,values=df.Age,aggfunc=np.average)
```
