```python
import pandas as pd
```

```python
india_weather = pd.DataFrame({
    "city" : ["mumbai","delhi","bangalore"],
    "temp" : [32,45,30],
    "humidity" : [80,60,78]
})

india_weather
```

```python
us_weather = pd.DataFrame({
    "city" : ["new_york","chicago","paris"],
    "temp" : [21,14,35],
    "humidity" : [68,65,75]
})

us_weather
```

```python
# concat both DataFrames:

df = pd.concat([india_weather,us_weather])
df
```

```python
# modify index from df:
df = pd.concat([india_weather,us_weather],ignore_index=True)
df
```

```python
# Display weather according to country:

df = pd.concat([india_weather,us_weather],keys=["india","usa"])
df
```

```python
df.loc["india"]
```

```python
temp_df = pd.DataFrame({
    "city" : ["mumbai","delhi","bangalore"],
    "temp" : [32,45,30]
},index=[0,1,2])

temp_df
```

```python
wind_df= pd.DataFrame({
    "city" : ["delhi","bangalore"],
    "windspeed" : [6,12]
},index=[1,2])

wind_df
```

```python
df = pd.concat([temp_df,wind_df])
df
```

```python
df = pd.concat([temp_df,wind_df],axis=1)
df
```

```python
# Add a column to temp_df and give appropriate values to it:

eve = pd.Series(["humid","dry","rain"],name="event")
eve
```

```python
df = pd.concat([temp_df,eve],axis=1)
df
```
