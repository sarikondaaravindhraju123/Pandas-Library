```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "city" : ["newyork","paris","virginia"],
    "temp" : [21,14,18]
})

df1
```

```python
df2 = pd.DataFrame({
    "city" : ["newyork","paris","virginia"],
    "humidity" : [65,68,75]
})

df2
```

```python
# on = 'city we use this because we merge using city:

df3 = pd.merge(df1,df2,on='city')
df3

# This is exactly join operation in mysql
```

```python
# case : what if two dfs have different cities we merge them using cities:
```

```python
df1 = pd.DataFrame({
    "city" : ["newyork","paris","virginia"],
    "temp" : [21,14,18]
})

df1
```

```python
df2 = pd.DataFrame({
    "city" : ["newyork","paris","texas","alaska"],
    "humidity" : [65,68,75,83]
})

df2
```

```python
df3 = pd.merge(df1,df2,on='city')
df3
```

```python
# join DataFrames and make it as outer join[sql] or union in  setr theory

df3 = pd.merge(df1,df2,on='city',how='outer')
df3
```

```python
# Left and right joins will be described by order of dfs u give for merge: 
```

```python
# Left outer join:

df3 = pd.merge(df1,df2,on='city',how='left')
df3
```

```python
# Right outer join:

df3 = pd.merge(df1,df2,on='city',how='right')
df3
```

```python
# indicate each of the city either it is left or right:

df3 = pd.merge(df1,df2,on='city',how='outer',indicator='True')
df3
```

```python
df1 = pd.DataFrame({
    "city" : ["newyork","paris","orlando","baltimore"],
    "temp" : [21,14,35,38],
    "humidity" : [65,68,71,75]
})

df1
```

```python
df2 = pd.DataFrame({
    "city" : ["chicago","newyork","san diego"],
    "temp" : [21,14,35],
    "humidity" : [65,68,71]
})

df2
```

```python
df3 = pd.merge(df1,df2,on='city',suffixes=['_left','_right'])
df3
```

```python
df3 = pd.merge(df1,df2,on='city',how='outer',suffixes=['_left','_right'])
df3
```
