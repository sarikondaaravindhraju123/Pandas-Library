```python
import pandas as pd
# creating df using python lists
weather_data = {
    "Day" : ["1/1/2017","1/2/2017","1/3/2017"],
    "Temp" : [32,42,58],
    "wind speed" :[30,40,50],
    "event" :["Rain","Sunny","Snow"]
}
df = pd.DataFrame(weather_data,columns = ["Day","Temp","wind speed","event"])
df
```

```python
# creating dataframe using tuples:
weather_data1 = [
    ("1/1/2017",32,6,"Rain"),
    ("1/2/2017",35,7,"Sunny"),
    ("1/3/2017",28,2,"Snow")
]
df = pd.DataFrame(weather_data1,columns=["Day","Temp","Wind speed","Event"])
df
```

```python
# creating df using dictionaries
weather_data2 = [
    {"Day":"1/1/2017","Temp":32,"Wind speed":6,"Event":"Rain"},
    {"Day":"1/2/2017","Temp":35,"Wind speed":7,"Event":"Sunny"},
    {"Day":"1/3/2017","Temp":28,"Wind speed":2,"Event":"Snow"}
]
df = pd.DataFrame(weather_data2)
df
```
