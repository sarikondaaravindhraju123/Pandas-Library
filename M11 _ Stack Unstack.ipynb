```python
import pandas as pd
```

```python
df = pd.read_excel("stocks (1).xlsx",header=[0,1])
df
```

```python
# To get rid of unnamed names we use index_col = 0:
df = pd.read_excel("stocks (1).xlsx",header=[0,1],index_col=[0])
df
```

```python
# stack(): It "stacks" the columns of the DataFrame into a single column, creating a new DataFrame with a hierarchical index.
# The resulting DataFrame has a row for each combination of the original row index and column index.

df.stack()
```

```python
df_stacked = df.stack(level=0)
df_stacked
```

```python
# we will get our df back by using unstack:

df_stacked.unstack()
```
