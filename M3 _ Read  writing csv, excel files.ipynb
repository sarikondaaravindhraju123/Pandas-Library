```python
# reading of excel file:
df = pd.read_excel("stock_data.xlsx")
df
```

```python
# converters : These we use to convert something into meaningfully:
# This is not a good way to do because if there are many companies lime RIL is "n.a" this will also return name as "sam walton"

def conv_people(cell):
  if cell == "n.a." or cell == "not available":
    return "sam Walton"
  return cell

def conv_eps(cell):
  if cell == "not available" or cell == "n.a":
    return None
  return cell

df = pd.read_excel("stock_data.xlsx",converters = {"people":conv_people,
                                                   "eps":conv_eps})
df
```

```python
# writing excel:
# skip_blank_lines : this is going to skip blank lines if u pass this when there are any blanks in the table:
# This above is only used in .csv not in .excel 
```

```python
df.to_excel("new.xlsx",sheet_name="stocks",index = False,startrow =1,startcol=2)
df  = pd.read_excel("new.xlsx","stocks",header=None)
df
```

```python
# write these two dataframes into same excel file as 2 different sheets
df_stocks = pd.DataFrame({
    "tickers" : ["Gogl","WMP","MSFT"],
    "price" : [84,45,64],
    "pe" : [30.347,14.26,30.97],
    "eps" : [27.82,4.61,2.12]
})

df_weather = pd.DataFrame({
    "day" : ["1/1/2017","1/2/2017","1/3/2017"],
    "temp" : [32,35,28],
    "event" : ["Rain","Sunny","Snow"]
})

# Excel we can write data frames into excel with thw help of ExcelWriter:
with pd.ExcelWriter("stocks_weather.xlsx") as writer:
  df_stocks.to_excel(writer,sheet_name="stocks"),
  df_weather.to_excel(writer,sheet_name="weather")

df = pd.read_excel("stocks_weather.xlsx","stocks") 
df
```
