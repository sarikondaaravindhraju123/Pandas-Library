```python
# pivot_table : It is used to summarize and aggregate data inside DataFrame:
```

```python
df = pd.read_csv("weather2.csv")
df
```

```python
# Above DataFrames consists of 2 different temps,humiditys in a day:
# So make avgs of temps,hunmidity and assign values:
```

```python
df.pivot_table(index="city",columns="date")
```

```python
# This aggregate function is ging to mean the values and return:
df.pivot_table(index="city",columns="date",aggfunc="mean")
```

```python
# This is going to do average by row and column wise:
df.pivot_table(index="city",columns="date",aggfunc="mean",margins="True")
```
